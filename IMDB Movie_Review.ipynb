```python
# Import necessary libraries
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/IMDB-Movie-Data.csv')  # Make sure this path matches where your dataset is located

# Top 10 Rated Comedy Movies
comedy_movies = data[data['Genre'].str.contains('Comedy', na=False)]  # Filter movies with 'Comedy' in Genre
top_comedy_movies = comedy_movies.sort_values(by='Rating', ascending=False).head(10)  # Sort by Rating and get top 10
print("Top 10 Rated Comedy Movies:")
print(top_comedy_movies[['Title', 'Rating']])

# Top 5 Rated Movies Overall
top_movies = data.sort_values(by='Rating', ascending=False).head(5)  # Sort by Rating and get top 5
print("Top 5 Rated Movies Overall:")
print(top_movies[['Title', 'Rating']])
```
